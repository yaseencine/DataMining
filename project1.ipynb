# AI Adoption and Job Automation Analysis

This notebook analyzes the relationship between AI adoption levels and the risk of job automation in various industries using visualizations and exploratory data analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

# Load the dataset
data = pd.read_csv("ai_job_market_insights.csv")
data.head()
```

## Preprocessing

```python
# Convert categorical features to numerical
data['AI_Adoption_Level'] = data['AI_Adoption_Level'].map({'Low': 1, 'Medium': 2, 'High': 3})
data['Automation_Risk'] = data['Automation_Risk'].map({'Low': 1, 'Medium': 2, 'High': 3})

# Check for nulls
data.isna().sum()
```

## Visualization 1: AI Adoption vs. Automation Risk

```python
sns.boxplot(x='AI_Adoption_Level', y='Automation_Risk', data=data)
plt.title('AI Adoption vs. Automation Risk')
plt.xlabel('AI Adoption Level')
plt.ylabel('Automation Risk')
plt.show()
```

## Visualization 2: Salary Range by AI Adoption and Automation Risk

```python
sns.violinplot(x='AI_Adoption_Level', y='Salary_USD', hue='Automation_Risk', data=data, split=True)

plt.title('Salary Range by AI Adoption and Automation Risk')
plt.xlabel('AI Adoption Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Automation Risk')
plt.show()
```

## Visualization 3: Growth Projections by AI Adoption Level

```python
sns.barplot(x='AI_Adoption_Level', y='Job_Growth_Projection', data=data)
plt.title('Growth Projections by AI Adoption Level')
plt.xlabel('AI Adoption Level')
plt.ylabel('Growth Projection')
plt.show()
```

## Key Takeaways

- Higher AI adoption levels often correspond to higher automation risks.
- Median salaries remain stable, but higher AI adoption shows narrower ranges with fewer outliers.
- Growth projections are more positive with low AI adoption and become increasingly negative with higher AI adoption.
